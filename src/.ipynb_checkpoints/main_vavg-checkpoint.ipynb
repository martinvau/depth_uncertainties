## Calculating depthing uncertainty for a 2-layer example 

### Problem
Assess *depth-conversion* uncertainties for a 2-layer velocity model related to 

a) TWT picking uncertainties and 

b) unconstrained seismic velocities of the second layer. 

```python
%matplotlib inline

# Libraries
from IPython.display import Image
from IPython.core.display import HTML 

import matplotlib.pyplot as plt
import matplotlib.cm as cm

import numpy as np
import math
```

```python
## User Input
z1 = 5000                    #depth surface1
vavg1 = v1 = 3750            # VAVG at depth z1

vint2_min = v2_min = 4500    # minimum interval velocity of second layer. default: 4500 m/s
vint2_max = v2_max = 6000    # maximum interval velocity of second layer. default: 6000 m/s
v2_inc = 250                 # velocity increment

rel_diff_twt1_twt2_min = 0.5 # min. relative difference between twt1 and twt2 pick in seconds. default: 0.5s = 500ms
rel_diff_twt1_twt2_max = 3.1 # max. relative difference between twt1 and twt2 pick in seconds. default: 1.5s = 1500ms
picking_resolution = 0.02    # picking increment in seconds. default: 0.02s = 20ms
```

```python
## Functions

def calc_vavg(depth1, vel1, twt2, vel2):
    '''
    Calculate Vavg at twt2 with given properties of first layer
    INPUT:  depth1: Absolute elevation depth Z1 of BOTTOM Layer1 [m] > 0
            vel1:   Average velocity at depth1 [m/s]
            twt2:   Absolute picking time TWT2 of BOTTOM Layer2 [s] > 0
            vel2:   Interval velocity V2 of Layer2 [m/s]
    OUTPUT: vout:   Avergae velocity at BOTTOM Layer2    
    '''
    
    twt1 = 2 * depth1 / vel1 #calculate TWT time of BOTTOM Layer1
    
    if twt2 > 20:
        raise ValueError("\tSI unit of twt2 most likely not in seconds. exit.")
        return 1
    
    if twt1 > twt2:
        raise ValueError("\thorizon 2 is above horizon 1. \n\t\ttwt1 = 2 * depth1 / vel1 * 1000 > twt2.\n\t\texit.")
        return 1 
    
    
    vout = ((2 * depth1) + (vel2 * (twt2-twt1))) / twt2 
    
    return vout
```

```python
# Calculations from user input

# establish TWT from BOTTOM od surface 1
twt1 = 2 * z1 / vavg1

vint2_max += v2_inc
v2_max += v2_inc

# create lists for possible velocities of v2 and possible TWT2 picking times
v2_vels = np.arange(v2_min,v2_max,v2_inc) # list of possible 
twt2_times = twt1 + np.arange(rel_diff_twt1_twt2_min,rel_diff_twt1_twt2_max,picking_resolution) # simulate list of possible absolute twt2 picks in [s]


```

```python
# calculate Vavg for each v2/twt2 pair and then corresponding absolute z2 depth

i = 0

z2_list = []                                      # list of absolute z2 values for all velocities

for vel in v2_vels:
    z2_temp = []
    for twt in twt2_times:

        vavg2 = calc_vavg(z1,vavg1,twt,vel)       # get Vavg at bottom of layer 2
        z2_temp.append(vavg2 * twt / 2)           # calculate resulting z2 depth from twt2 and vel2 and store in list
    
    z2_list.append(z2_temp)
    i += 1
    
```

```python
## calculate relative values
twt2_times_rel = [x - twt1 for x in twt2_times]


z2_depths_rel = []

for x in range(len(z2_list)):
    z2_depths_rel.append([x - z1 for x in z2_list[x]])
```

### Where to find what:
* `twt2_times`: Absolute picking times for twt2, calculated from `twt1` and **user input**

* `twt2_times_rel`: Relative picking times for twt2 with respect to `twt1`. Range from **user input**

* `z2_list`: Absolute depths for `z2`. Calculated from `vavg2` and absolute twt2 from **user input**. The indices correspond to the `v2_int` velocities in `v2_vels`, i.e. Depth list in `z2_list[0]`are a function of `twt2_times` array and `v2_vels[0]`. 

* `z2_depths_rel`: Relative depths for `z2` *below* `z1` (= stratigraphic thickness). Calculated from `vavg2` and absolute twt2 from **user input**. The indices correspond to the `v2_int` velocities in `v2_vels`, i.e. Depth list in `z2_list[0]`are a function of `twt2_times` array and `v2_vels[0]`. 



```python
# Initiate plot
fig, ax = plt.subplots(figsize=(10, 10)) 



## Format plot
color_list = ['r','g','b','k','pink','c','orange','purple']

## Titles ---------------------------------------
ttl = fig.suptitle('Time vs Depth Stratigraphic Thickness in Dependence of Seismic Velocities', 
             fontsize=18, fontweight='bold')
ttl.set_position([.5, 0.95])

ax.set_title('Thickness overburden: '+ str(z1) +'m, Vavg overburden: ' + str(vavg1) + 'm/s',
            fontsize=16) 

## Axes labels and ticks ---------------------------------------
ax.set_ylabel('Resulting relative stratigrpahic thickness [m]', fontsize=16) 
ax.set_xlabel('Relative stratigraphic time thickness [s]', fontsize=16) 

ax.tick_params(direction='in', length=12, width=2, colors='k',
               grid_color='gray', grid_alpha=0.5)


### Automatically adjust min/max axes values
plt.xticks(np.arange(min(twt2_times_rel), max(twt2_times_rel)+2, 0.2))
min_y = math.floor(min(z2_depths_rel[0]) / 1000) * 1000
plt.yticks(np.arange(min_y, max(z2_depths_rel[(len(v2_vels)-1)])+250, 500))




## Grid -----------------------------------------------
plt.grid(color='gray', linestyle='--', linewidth=1)

## plot data ------------------------------------------

for m in range(len(z2_depths_rel)):
    legend_label = str(v2_vels[m]) + ' m/s'
    ax.plot(twt2_times_rel, z2_depths_rel[m], color_list[m], label=legend_label)


ax.legend()

plt.show()
```
